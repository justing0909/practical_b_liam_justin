```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame
data = {
    "nomic-embed-text": [136.9043, 163.8902, 158.9918, 196.5097, 160.4392, 170.0077],
    "all-minilm": [170.0637, 167.5921, 179.9499, 166.0753, 160.7731, 169.9198],
    "snowflake-arctic-embed": [157.2397, 229.8122, 183.1776, 182.3144, 177.3967, 183.8753]
}

df = pd.DataFrame(data, index=["Sample 1", "Sample 2", "Sample 3", "Sample 4", "Sample 5", "Sample 6"])

# Create the box plot with filled boxes
plt.figure(figsize=(8, 6))
box = df.boxplot(grid=True, patch_artist=True)  # patch_artist=True fills the boxes

# Customize colors
colors = ['lightblue', 'lightgreen', 'lightcoral']

# Labels and title
plt.xlabel("Embedding Models")
plt.ylabel("Ingest Time (Seconds)")
plt.title("Box and Whiskers Plot of Embedding Models Ingest Time")

# Show the plot
plt.show()
```
